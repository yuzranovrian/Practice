```python
data_toko = {
    "Produk": ["Kopi", "Teh", "Susu"],
    "Harga": [18000, 15000, 20000],
    "Stok": [10, 15, 8]
}

df = pd.DataFrame(data_toko)

df
```

```python
df["Total_Nilai"] = df["Harga"] * df["Stok"]

df
```

```python
stok_tipis = df[df["Stok"] < 10]

stok_tipis
```

```python
df.to_excel("Laporan_Toko.xlsx", index=False)

print("File Excel Berhasil Dibuat!")
```
